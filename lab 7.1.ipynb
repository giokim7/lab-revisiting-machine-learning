```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('learningSet.csv', low_memory=False)
```

```python
#check null values
null_values_df =df.isna().sum()
print(null_values_df)
```

```python
print(df.columns)
```

```python
def get_drop_columns(df, threshold = 0.85):
    nulls_percent_df = pd.DataFrame(df.isna().sum()/len(df)).reset_index()
    nulls_percent_df.columns = ['column_name', 'nulls_percentage']
    columns_above_threshold = nulls_percent_df[nulls_percent_df['nulls_percentage']>threshold]
    drop_columns_list = list(columns_above_threshold['column_name'])
    return drop_columns_list
```

```python
drop_columns_list = get_drop_columns(df, threshold=0.85)
print(drop_columns_list)
```

```python
drop_columns_list.extend(['OSOURCE', 'ZIP'])
```

```python
df.drop(columns=drop_columns_list, inplace=True)
```

```python
#replace values in gender column
print(df['GENDER'].value_counts())
df['GENDER'] = df['GENDER'].fillna('F')
```
